```python
# --- Inicialización de variables y listas ---
match = re.match(r"(\d+(?:\.\d+)?)S", os.path.basename(base_dir))
if match:
    total_time = float(match.group(1))
    print(f"Total time encontrado: {total_time}")
else:
    print("No se encontró un número seguido de 'S' en el nombre de carpeta.")

labels = []
stft_means_f, wavelet_means_f = [], []
stft_means_t, wavelet_means_t = [], []
stft_stds_f, wavelet_stds_f = [], []
stft_stds_t, wavelet_stds_t = [], []

superlet1_means_f, superlet2_means_f, superlet3_means_f = [], [], []
superlet1_means_t, superlet2_means_t, superlet3_means_t = [], [], []
superlet1_stds_f, superlet2_stds_f, superlet3_stds_f = [], [], []
superlet1_stds_t, superlet2_stds_t, superlet3_stds_t = [], [], []

tonalidades = [0.4, 0.6, 0.8]

colores_base = {                                 
    "STFT": plt.colormaps["Blues"],              
    "Wavelet": plt.colormaps["Greens"],          
    "Superlet": plt.colormaps["Oranges"],        
}              

# --- Función auxiliar ---
def extract_frequency(folder_name):
    match = re.search(r"([\d\.]+)\s*Hz", folder_name)
    return float(match.group(1)) if match else float('inf')

# --- Filtrar y ordenar carpetas ---
sorted_folders = sorted([f for f in os.listdir(base_dir) if "Hz" in f], key=extract_frequency)

# --- Recorrer carpetas y cargar datos ---
for folder in sorted_folders:
    folder_path = os.path.join(base_dir, folder)

    if os.path.isdir(folder_path):
        filename_1 = os.path.join(folder_path, "maes_f.txt")
        filename_2 = os.path.join(folder_path, "maes_t.txt")

        if os.path.exists(filename_1) and os.path.exists(filename_2):
            with open(filename_1, "r", encoding="utf-8") as file:
                maes_f = ast.literal_eval(file.read())

            with open(filename_2, "r", encoding="utf-8") as file:
                maes_t = ast.literal_eval(file.read())

            labels.append(folder)

            # STFT y Wavelet
            stft_means_f.append(maes_f["stft"][1][1])
            wavelet_means_f.append(maes_f["wavelet"][1][1])
            stft_stds_f.append(maes_f["std_stft"][1][1])
            wavelet_stds_f.append(maes_f["std_wavelet"][1][1])
            
            stft_means_t.append(maes_t["stft"][1][1])
            wavelet_means_t.append(maes_t["wavelet"][1][1])
            stft_stds_t.append(maes_t["std_stft"][1][1])
            wavelet_stds_t.append(maes_t["std_wavelet"][1][1])

            # Superlet (3 parámetros)
            superlet1_means_f.append(maes_f["superlet"][1][0])
            superlet2_means_f.append(maes_f["superlet"][1][1])
            superlet3_means_f.append(maes_f["superlet"][1][2])
            superlet1_stds_f.append(maes_f["std_superlet"][1][0])
            superlet2_stds_f.append(maes_f["std_superlet"][1][1])
            superlet3_stds_f.append(maes_f["std_superlet"][1][2])

            superlet1_means_t.append(maes_t["superlet"][1][0])
            superlet2_means_t.append(maes_t["superlet"][1][1])
            superlet3_means_t.append(maes_t["superlet"][1][2])
            superlet1_stds_t.append(maes_t["std_superlet"][1][0])
            superlet2_stds_t.append(maes_t["std_superlet"][1][1])
            superlet3_stds_t.append(maes_t["std_superlet"][1][2])

# --- Preparar coordenadas ---
x = np.arange(len(labels))
width = 0.17

# --- Crear figura con subplots compartiendo eje x ---
fig, (ax_f, ax_t) = plt.subplots(2, 1, figsize=(12,13.5), sharex=True, gridspec_kw={'height_ratios':[1,1]})

# --- Primer subplot: MAE frecuencia ---
ax_f.bar(x - 2*width, stft_means_f, width, yerr=stft_stds_f, capsize=5, label="STFT ws: 634 ms", color=colores_base["STFT"](tonalidades[1]))
ax_f.bar(x - width, wavelet_means_f, width, yerr=wavelet_stds_f, capsize=5, label="CWT c: 130", color=colores_base["Wavelet"](tonalidades[1]))
ax_f.bar(x, superlet1_means_f, width, yerr=superlet1_stds_f, capsize=5, label="SLT c1:2, o:3-220", color=colores_base["Superlet"](tonalidades[0]))
ax_f.bar(x + width, superlet2_means_f, width, yerr=superlet2_stds_f, capsize=5, label="SLT c1:3, o:4-250", color=colores_base["Superlet"](tonalidades[1]))
ax_f.bar(x + 2*width, superlet3_means_f, width, yerr=superlet3_stds_f, capsize=5, label="SLT c1:6, o:7-300", color=colores_base["Superlet"](tonalidades[2]))

ax_f.set_ylabel("$\mathregular{MAE_f}$ (Hz)", fontsize=fontsize_labels)
ax_f.set_ylim(0, 17.5)
ax_f.set_yticks(np.arange(0, 20, 2.5))
ax_f.set_yticklabels(np.arange(0, 20, 2.5), fontsize=fontsize_ticks) 
ax_f.grid(axis="y", linestyle="--", alpha=0.7)
# ax_f.legend(loc='upper right', fontsize=fontsize_legend)
# ax_f.set_title("MAE Comparison - Frequency & Time", fontsize=14)

# --- Segundo subplot: MAE tiempo ---
ax_t.bar(x - 2*width, stft_means_t, width, yerr=stft_stds_t, capsize=5, label="STFT ws: 634 ms", color=colores_base["STFT"](tonalidades[1]))
ax_t.bar(x - width, wavelet_means_t, width, yerr=wavelet_stds_t, capsize=5, label="CWT c:130", color=colores_base["Wavelet"](tonalidades[1]))
ax_t.bar(x, superlet1_means_t, width, yerr=superlet1_stds_t, capsize=5, label="SLT c1:2, o:3-220", color=colores_base["Superlet"](tonalidades[0]))
ax_t.bar(x + width, superlet2_means_t, width, yerr=superlet2_stds_t, capsize=5, label="SLT c1:3, o:4-250", color=colores_base["Superlet"](tonalidades[1]))
ax_t.bar(x + 2*width, superlet3_means_t, width, yerr=superlet3_stds_t, capsize=5, label="SLT c1:6, o:7-300", color=colores_base["Superlet"](tonalidades[2]))

ax_t.set_xlabel("Ideal MNF", fontsize=fontsize_labels)
ax_t.set_ylabel("$\mathregular{MAE_t}$ (s)", fontsize=fontsize_labels)
ax_t.set_ylim(0, 1)
ax_t.set_yticks(np.arange(0, 1.1, 0.1))
ax_t.set_yticklabels([f"{y:.1f}" for y in np.arange(0, 1.1, 0.1)], fontsize=fontsize_ticks) 
ax_t.grid(axis="y", linestyle="--", alpha=0.7)
ax_t.legend(loc="upper right", fontsize=fontsize_legend)
ax_t.set_xticks(x)
ax_t.set_xticklabels(labels, rotation=45, ha="right", fontsize=fontsize_labels)

# --- Eje derecho para porcentaje de MAE_t ---
ax_right = ax_t.twinx()
ax_right.set_ylim(0, 1)
ax_right.set_yticks(np.arange(0, 1.1, 0.1))
ax_right.set_yticklabels([f"{(y / total_time * 100):.0f}%" for y in np.arange(0, 1.1, 0.1)], fontsize=fontsize_labels)
ax_right.set_ylabel(f"MAE (% of {total_time:.2f} s)", fontsize=fontsize_labels)

# --- Ajuste final y guardar ---
plt.tight_layout()
fig.savefig(os.path.join(base_dir, "MAE_comparison_combined.png"), bbox_inches='tight')
plt.show()
```

```python
import os
import re
import ast
import numpy as np
import matplotlib.pyplot as plt

# Inicializar listas para almacenar datos
# Buscar número antes de 'S' al inicio de la cadena
match = re.match(r"(\d+(?:\.\d+)?)S", os.path.basename(base_dir))
if match:
    total_time = float(match.group(1))
    print(f"Total time encontrado: {total_time}")
else:
    print("No se encontró un número seguido de 'S' en el nombre de carpeta.")

labels = []
stft_means_f, wavelet_means_f = [], []
stft_means_t, wavelet_means_t = [], []
stft_stds_f, wavelet_stds_f = [], []
stft_stds_t, wavelet_stds_t = [], []

# Inicializar listas para Superlet (3 parámetros diferentes)
superlet1_means_f, superlet2_means_f, superlet3_means_f = [], [], []
superlet1_means_t, superlet2_means_t, superlet3_means_t = [], [], []
superlet1_stds_f, superlet2_stds_f, superlet3_stds_f = [], [], []
superlet1_stds_t, superlet2_stds_t, superlet3_stds_t = [], [], []

tonalidades = [0.4, 0.6, 0.8]

colores_base = {                                 
    "STFT": plt.colormaps["Blues"],              
    "Wavelet": plt.colormaps["Greens"],          
    "Superlet": plt.colormaps["Oranges"],        
}              

# Función para extraer frecuencia del nombre de carpeta
def extract_frequency(folder_name):
    match = re.search(r"([\d\.]+)\s*Hz", folder_name)
    return float(match.group(1)) if match else float('inf')

# Filtrar y ordenar carpetas por frecuencia
sorted_folders = sorted([f for f in os.listdir(base_dir) if "Hz" in f], key=extract_frequency)

# Recorrer todas las carpetas
for folder in sorted_folders:
    folder_path = os.path.join(base_dir, folder)

    if os.path.isdir(folder_path):
        filename_1 = os.path.join(folder_path, "maes_f.txt")
        filename_2 = os.path.join(folder_path, "maes_t.txt")

        if os.path.exists(filename_1) and os.path.exists(filename_2):
            with open(filename_1, "r", encoding="utf-8") as file:
                maes_f = ast.literal_eval(file.read())

            with open(filename_2, "r", encoding="utf-8") as file:
                maes_t = ast.literal_eval(file.read())

            # Guardar etiquetas
            labels.append(folder)

            # STFT y Wavelet
            stft_means_f.append(maes_f["stft"][1][-2])
            wavelet_means_f.append(maes_f["wavelet"][1][-2])
            stft_stds_f.append(maes_f["std_stft"][1][-2])
            wavelet_stds_f.append(maes_f["std_wavelet"][1][-2])
            
            stft_stds_t.append(maes_t["std_stft"][1][-2])
            wavelet_stds_t.append(maes_t["std_wavelet"][1][-2])

            stft_means_t.append(maes_t["stft"][1][-2])
            wavelet_means_t.append(maes_t["wavelet"][1][-2])

            # Superlet (tres parámetros)
            superlet1_means_f.append(maes_f["superlet"][1][0])
            superlet2_means_f.append(maes_f["superlet"][1][1])
            superlet3_means_f.append(maes_f["superlet"][1][2])

            superlet1_stds_f.append(maes_f["std_superlet"][1][0])
            superlet2_stds_f.append(maes_f["std_superlet"][1][1])
            superlet3_stds_f.append(maes_f["std_superlet"][1][2])
            
            superlet1_means_t.append(maes_t["superlet"][1][0])
            superlet2_means_t.append(maes_t["superlet"][1][1])
            superlet3_means_t.append(maes_t["superlet"][1][2])
        
            superlet1_stds_t.append(maes_t["std_superlet"][1][0])
            superlet2_stds_t.append(maes_t["std_superlet"][1][1])
            superlet3_stds_t.append(maes_t["std_superlet"][1][2])


# Ancho de las barras
x = np.arange(len(labels))
width = 0.17

# Crear figura para frecuencia
fig, ax = plt.subplots(figsize=(10.5,5))

# Gráfico de frecuencia
ax.bar(x - 2*width, stft_means_f, width, yerr=stft_stds_f, capsize=5, label="STFT ws: 800 ms", color=colores_base["STFT"](tonalidades[1]))
ax.bar(x - width, wavelet_means_f, width, yerr=wavelet_stds_f, capsize=5, label="CWT c: 150", color=colores_base["Wavelet"](tonalidades[1]))
ax.bar(x, superlet1_means_f, width, yerr=superlet1_stds_f, capsize=5, label="SLT $\mathregular{c_1}$:5, o:1-100", color=colores_base["Superlet"](tonalidades[0]))
ax.bar(x + width, superlet2_means_f, width, yerr=superlet2_stds_f, capsize=5, label="SLT $\mathregular{c_1}$:8, o:4-150", color=colores_base["Superlet"](tonalidades[1]))
ax.bar(x + 2*width, superlet3_means_f, width, yerr=superlet3_stds_f, capsize=5, label="SLT $\mathregular{c_1}$:10, o:6-200", color=colores_base["Superlet"](tonalidades[2]))

ax.set_xlabel("Ideal MNF", fontsize=fontsize_labels)
ax.set_ylabel("$\mathregular{MAE_f}$ (Hz)", fontsize=fontsize_labels)
# ax.set_title("Comparison of MAE in frequency according to the MNF of interest", fontsize=fontsize_title)
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45, ha="right", fontsize=fontsize_ticks-1)
ax.legend(loc='upper right',fontsize=fontsize_legend)
ax.set_yticks(np.arange(0, 25, 2.5))
ax.grid(axis="y", linestyle="--", alpha=0.7)
ax.set_ylim(0,22.5)


# Crear la figura y eje izquierdo
fig_2, ax_2 = plt.subplots(figsize=(9, 5.5))

# Dibujar las barras en el eje izquierdo
ax_2.bar(x - 2*width, stft_means_t, width, yerr=stft_stds_t, capsize=5, label="STFT ws: 800 ms", color=colores_base["STFT"](tonalidades[1]))
ax_2.bar(x - width, wavelet_means_t, width, yerr=wavelet_stds_t, capsize=5, label="CWT c:150", color=colores_base["Wavelet"](tonalidades[1]))
ax_2.bar(x, superlet1_means_t, width, yerr=superlet1_stds_t, capsize=5, label="SLT $\mathregular{c_1}$:5, o:1-100", color=colores_base["Superlet"](tonalidades[0]))
ax_2.bar(x + width, superlet2_means_t, width, yerr=superlet2_stds_t, capsize=5, label="SLT $\mathregular{c_1}$:8, o:4-150", color=colores_base["Superlet"](tonalidades[1]))
ax_2.bar(x + 2*width, superlet3_means_t, width, yerr=superlet3_stds_t, capsize=5, label="SLT $\mathregular{c_1}$:10, o:6-200", color=colores_base["Superlet"](tonalidades[2]))

# 🔧 Corrección: obtener y filtrar yticks positivos
ax_2.set_ylim(0, 1)

# 🔧 Crear eje derecho sincronizado solo después de fijar límites
ax_right = ax_2.twinx()
ax_right.set_ylim(0, 1)

# Etiquetas del eje derecho como porcentaje
yticks = np.arange(0, 1.1, 0.1)  # [0.0, 0.2, ..., 1.0]
ax_2.set_yticks(yticks)
ax_right.set_yticks(yticks)
ax_right.set_yticklabels([f"{(y / total_time * 100):.0f}%" for y in yticks])

ax_right.set_ylabel(f"MAE  (% of {total_time:.2f} s)", fontsize=fontsize_labels)

# Configuración general
ax_2.set_xlabel("Ideal MNF", fontsize=fontsize_labels)
ax_2.set_ylabel("$\mathregular{MAE_t}$ (s)", fontsize=fontsize_labels)
# ax_2.set_title("Comparison of MAE in time according to the MNF of interest", fontsize=fontsize_title)
ax_2.set_xticks(x)
ax_2.set_xticklabels(labels, rotation=45, ha="right", fontsize=fontsize_labels-1)
ax_2.grid(axis="y", linestyle="--", alpha=0.7)
ax_2.legend(loc="upper right", fontsize=fontsize_legend)

# Guardar y mostrar
fig.savefig(os.path.join(base_dir, "MAE_frequency_comparison.png"), bbox_inches='tight')
fig_2.savefig(os.path.join(base_dir, "MAE_time_comparison.png"), bbox_inches='tight')
plt.tight_layout()
plt.show()
```
